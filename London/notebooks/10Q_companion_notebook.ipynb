```python
random_state = 1
elec_1house = elec_house.sample(n=1, random_state=random_state, axis=1)
elec_10house = elec_house.sample(n=10, random_state=random_state, axis=1).sum(axis=1).to_frame()
elec_100house = elec_house.sample(n=100, random_state=random_state, axis=1).sum(axis=1).to_frame()
elec_1000house = elec_house.sample(n=1000, random_state=random_state, axis=1).sum(axis=1).to_frame()

elec_1house.columns = ['kWh']
elec_1house['Number of houses'] = 1
elec_10house.columns = ['kWh']
elec_10house['Number of houses'] = 10
elec_100house.columns = ['kWh']
elec_100house['Number of houses'] = 100
elec_1000house.columns = ['kWh']
elec_1000house['Number of houses'] = 1000

elec = pd.concat([elec_1house,elec_10house,elec_100house,elec_1000house],axis=0).reset_index()
elec = elec.rename(columns={'tstp':'timestamp'})
elec = elec[(elec['timestamp']>='2013-06-01')&(elec['timestamp']<'2013-07-01')]

g = sns.FacetGrid(elec, row="Number of houses", #row_order=ordered_days,
                  height=1.5, aspect=5, sharey=False)
g.map(sns.lineplot, "timestamp", "kWh")
plt.savefig('agg_level.jpg',bbox_inches='tight', dpi=300)
```

## Understanding the drivers of electricity demand

As explained in the paper, several factors influence electricity demand in buildings. Some of the most important factors include:

1. Historic demand (aka the autoregressive component)
2. Calendar features (time of day, day of year etc.)
3. Ambient weather conditions (temperature, humidity, wind speed etc.)
4. Other exogenous factors (e.g. occupancy levels etc.)

Here, we focus on the first two of these factors, and show how they influence the electricity demand. In the next section, we will use these insights to build our first forecast models.

### Autoregressive component

The autoregressive component in a time series can be easily identified by its autocorrelation and partial autocorrelation function. More concretely, the autocorrelation function (ACF) identifies how a time series is related to its own past (the partial autocorrelation function or PACF removes the effect of interceding lags and tells us how strongly the time series depends on a specific lag of its own past). 

There are several implementations of ACF and PACF which can be used easily in Python. Here we make use of the `Statsmodel` implementation.

```python
df_plot = elec_city.resample('H').mean()

fig = plt.figure(figsize=(12,8))

ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df_plot['energy(kWh/hh)'], lags=24*7+1, ax=ax1)
ax1.xaxis.set_ticks_position('bottom')
fig.tight_layout();

ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df_plot['energy(kWh/hh)'], lags=24*7+1, ax=ax2, method='ywm')
ax2.xaxis.set_ticks_position('bottom')
fig.tight_layout();
```

Both the ACF and PACF identify strong serial and seasonal autocorrelation in the time series (note that we used the aggregated electricity demand in this case i.e. `elec_city`). Contrast this against the next plots where we use the electricity demand from a single household to obtain the ACF and PACF plots. The same overall trend is visible, but the strength of the autocorrelation (i.e. the y-axis values) is much weaker (showing dimished seasonality).

```python
df_plot = elec_1house.resample('H').mean().dropna()

fig = plt.figure(figsize=(12,8))

ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df_plot['kWh'], lags=24*7+1, ax=ax1)
ax1.xaxis.set_ticks_position('bottom')
fig.tight_layout();

ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df_plot['kWh'], lags=24*7+1, ax=ax2, method='ywm')
ax2.xaxis.set_ticks_position('bottom')
fig.tight_layout();
```

Based on these plots, it is obvious that, to forecast electricity demand either on building or urban scale, including historic lags is a good idea. While these lags can differ by individual building (this is something that can be identified in the plots in fact), it is quite obvious that the time series show strong daily (and weekly) seasonality which can be exploited to make predictions about the future.

### The effect of calendar features

Besides the timestamp accompanying each meter reading, we can also retrieve holiday/calendar data from the website: www.timeanddate.com, which gives calendar information of a given year for countries worldwide. Note that in the other 
    
    
    on predicting load for Belgium, we used a different library.

```python
df_holiday = []

for year in np.arange(2011, 2014+1):
    df_holiday_temp = pd.read_html('https://www.timeanddate.com/holidays/uk/'+str(year)+'?hol=9')[0]
    df_holiday_temp.columns = df_holiday_temp.columns.get_level_values(0)
    df_holiday_temp = df_holiday_temp.dropna(how='all')
    df_holiday_temp = df_holiday_temp[['Date', 'Name', 'Type']]
    
    df_holiday_temp['Date'] = str(year) + ' ' + df_holiday_temp['Date']
    df_holiday_temp['Date'] = pd.to_datetime(df_holiday_temp['Date'], errors='coerce')
    df_holiday_temp = df_holiday_temp.dropna(subset=['Date'])
    
    df_holiday_temp = df_holiday_temp[df_holiday_temp['Type'].str.contains('Holiday')]
    
    df_holiday.append(df_holiday_temp)
    
df_holiday = pd.concat(df_holiday, axis=0, ignore_index=True)
df_holiday = df_holiday.drop_duplicates(subset=['Date'])
df_holiday = df_holiday.set_index('Date').asfreq('D')
df_holiday.loc[df_holiday.index.weekday>=5, 'Name'] = 'Weekend'
df_holiday.loc[df_holiday.index.weekday>=5, 'Type'] = 'Weekend'
df_holiday = df_holiday.fillna('Regular')
df_holiday.columns = 'holiday_' + df_holiday.columns

df_holiday = df_holiday.reset_index()
df_holiday = df_holiday.rename(columns={'Date':'date'}) 
df_holiday.loc[df_holiday['holiday_Type'].str.contains('Holiday'), 'holiday_Type'] = 'Holiday'

df_holiday
```

Next, we can look into the breakdown of weekday and day types to understand how (if) electricity demand varies as a function of the time of day or day of week or year etc.

```python
# 1st plot
df_plot = elec_city.resample('D').mean()
df_plot = df_plot.merge(df_holiday.set_index('date'), left_index=True, right_index=True)
df_plot['weekday'] = df_plot.index.weekday
df_plot['weekday'] = df_plot['weekday'].replace({0:'Mon',1:'Tue',2:'Wed',3:'Thur',4:'Fri',5:'Sat',6:'Sun'})
plt.figure(figsize=(6*0.65,4*0.65))
sns.boxplot(data=df_plot, x='weekday', y='energy(kWh/hh)',order=['Mon','Tue','Wed','Thur','Fri','Sat','Sun'])
#plt.ylim(6000,12000)
plt.xlabel('')#('Day of week')
plt.ylabel('Load (daily mean)')

# 2nd plot
plt.figure(figsize=(3.65*0.65,4*0.65))
sns.boxplot(data=df_plot, x='holiday_Type', y='energy(kWh/hh)',order=['Regular','Weekend','Holiday'])
#plt.ylim(6000,12000)
plt.xlabel('')#('Day type')
plt.ylabel('Load (daily mean)')
```

## Forecasting model of electricity consumption

In this section, we finally turn our attention to looking into forecasting electricity demand at both the building and urban level.

### Organizing data as (predictive) features

Before building the forecasting model, we need to organize the dataset into a dataframe which contains the necessary features, including various lags, temporal features (e.g., weekday and hour) and ambient conditions. Note that in the last notebook we relied on Darts to perform this pre-processing step for us:

```python
df_dataset = elec_city.copy()
df_dataset = df_dataset.rename(columns={'energy(kWh/hh)':'kWh'})
df_dataset = df_dataset.resample('H').mean()

df_dataset['timestamp'] = df_dataset.index
df_dataset['date'] = pd.to_datetime(df_dataset.index.date)
df_dataset['year'] = df_dataset.index.year.astype('int')
df_dataset['weekday'] = df_dataset.index.weekday
df_dataset['hour'] = df_dataset.index.hour + df_dataset.index.minute/60
df_dataset['kWh_yesterday'] = df_dataset['kWh'].shift(24)
df_dataset['kWh_lastWeek'] = df_dataset['kWh'].shift(24*7)

df_dataset = df_dataset.merge(weather_hourly_darksky[['temperature']], left_index=True, right_index=True)

df_dataset.set_index('timestamp',inplace=True)

df_dataset = df_dataset.dropna()

df_dataset
```

```python
df_plot = df_dataset
sns.pairplot(df_plot)
```

This relationship can also be visualized as a correlation matrix or a correlogram, which summarizes the strength of relationship between the different variables in the dataset.

Based on these plots, it appears that the electricity demand (kWh) correlates very well with historic demand (kWh_yesterday and kWh_lastWeek). This low correlation value also shows one of the core limitations of using (Pearson's) correlation to select features.

## Building and evaluating forecast models

### Defining error metrics

There are several (countless, almost) variations on error metrics that researchers and practitioners have used to evaluate demand forecast models. In this notebook, we evaluate all models on the following metrics:

1. $R^2$ (coefficient of determination)
2. MAE (mean absolute error)
4. MSE (mean squared error)

Each of these metrics highlights a different aspect of the forecaster's performance, but many others are also used in literature including MAPE, rMAE, AIC and BIC etc. Refer to the 10 questions paper's main body for more details.

### Selecting between different forecast models

It is also possible to build many different types of forecast models. Here, we refer only to function approximation techniques by this, so we limit our attention to a few choices only. These include:

1. Linear regression 
2. Neural Network
3. Random Forest
4. Gradient Boosting Trees
5. LightGBM
6. CatBoost

It is important to try several different models of varying complexity to ensure that a more complex model is not chosen instead of a simpler one. To do so, recall the discussion about using naive or persistence models for benchmarking from the other notebook. We leave the implementation of these simple models to the interested reader.

### Building a forecaster

We show here how to build a forecaster for the aggregated dataset, and leave it as an exercise for the reader to extend it to the case of individual buildings. Keep in mind that the accuracy, keeping all else constant, is much higher for aggregations of buildings as opposed to individual buildings.

First, let's define the input features we will use to make the forecast.

```python
list_feat = ['weekday', 'hour','temperature', 'kWh_lastWeek','kWh_yesterday']
```

Second, let's define the models we will fit to make our prediction.

```python
models = [LinearRegression(), MLPRegressor(), RandomForestRegressor(), GradientBoostingRegressor(),
          LGBMRegressor(), CatBoostRegressor(verbose=0)]

models_name = ['LinearRegression','Neural Network', 'Random Forest', 
               'Gradient Boosting Trees', 'LightGBM', 'CatBoost']
```

Finally, let's list which metrics we will use to evaluate our model.

```python
r2_metric = []
mae_metric = []
mse_metric = []
```

Now we can train different models on the training dataset (observed electricity demand in 2012), and estimate their performance on the test set (data from 2013).

```python
for model, model_name in zip(models, models_name):
    
    # Define training and test data
    traindata = df_dataset.loc[:'2012'].copy()
    testdata = df_dataset.loc['2013':].copy()

    # Fit the model 
    model.fit(traindata[list_feat], traindata['kWh'], )
    
    # Make predictions on the defined test set
    testdata['kWh_pred'] = model.predict(testdata[list_feat])
    
    # Calculate error metrics
    RSQUARED = r2_score(testdata['kWh'], testdata['kWh_pred'])
    MAE = mean_absolute_error(testdata['kWh'], testdata['kWh_pred'])
    MSE = mean_squared_error(testdata['kWh'], testdata['kWh_pred'])

    # Append error metrics to the list
    r2_metric.append(RSQUARED)
    mae_metric.append(MAE)
    mse_metric.append(MSE)
    
    print(model_name+' / RSQUARED:'+str(round(RSQUARED,3))+' / MAE: '+str(round(MAE, 4))+ ' / MSE: '+str(round(MSE, 4)))
```

```python
df_metrics = pd.DataFrame([models_name, r2_metric, mae_metric, mse_metric], index=['model','R2','MAE','MSE']).T
df_metrics
```

Based on the metrics dataframe, it is obvious that most models perform remarkably similar on all three metrics with the MLP-based neural network being the only notable exception. It should also be kept in mind that we have not performed any hyperparameter search for these models which can lead to a substantial deterioration in performance.

### The effect of forecast horizon

In this section, we take a closer look at how the choice of forecast horizon affects forecast performance.

#### Day-ahead forecasting

A common forecasting horizon is day-ahead forecasting of energy demand, which can be used for short-term planning (note that this is a simplification as, for many markets, day-ahead forecasts need to be made 36 to 38 hours in advance. With a shorter forecasting window, the lag feature of values 24 hours ago could be therefore used as one of the input features. Using the same feature list as defined previously, here we visualize the forecast made by the Gradient Boosting Regressor. It is however trivial to replace this model with any of the other models used previously.

```python
traindata = df_dataset.loc[:'2012'].copy()
testdata = df_dataset.loc['2013':].copy()

GB_model = GradientBoostingRegressor()
GB_model.fit(traindata[list_feat], traindata['kWh'])
testdata['kWh_pred'] = GB_model.predict(testdata[list_feat])
errors = abs(testdata['kWh_pred'] - testdata['kWh'])
RSQUARED = r2_score(testdata['kWh'], testdata['kWh_pred'])
MAE = mean_absolute_error(testdata['kWh'], testdata['kWh_pred'])

print("Model: GradientBoostingRegressor")
print("R SQUARED: "+str(round(RSQUARED,4)))
print("MAE: "+str(round(MAE,4)))

plt.figure(figsize=(8,2))
plt.plot(testdata.iloc[:24].index, testdata.iloc[:24]['kWh'], '-k', alpha=0.6, label='kWh')
plt.plot(testdata.iloc[:24].index, testdata.iloc[:24]['kWh_pred'], '--g', alpha=0.8, label='kWh_pred')
plt.legend(loc='lower center', ncol=3, bbox_to_anchor=(0.5, -0.5))
plt.show() 
```

This forecast is reasonably accurate in that it manages to capture the shape reasonably well. However, it does end up over-predicting during the afternoon and early morning. Is this over-forecast an isolated event or a structural issue? We leave this question to the interested reader to explore by making several forecasts, and then calculating the ACF/PACF of the resulting residual signal.

#### Week-ahead forecasting

Another common task for medium-term energy management is to make a week-ahead forecast. This forecasting model is built on not only temporal and weather features but also the lag feature of load values 168 hours ago (it cannot use the last day's value out of the box (although it is possible to incorporate this value recursively).

```python
list_feat = ['weekday', 'hour','temperature', 'kWh_lastWeek']
```

Since tree-based model is one of dominant methods showcased in many competitions and research, also with its low requirement of computational power and processing time, this example is also based on the gradient boosting regressor in the sklearn package. With the span from November 2011 to February 2014, we train the model based on the data from 2011 to 2012, and test the model via the retained data from 2013 to 2014.

```python
traindata = df_dataset.loc[:'2012'].copy()
testdata = df_dataset.loc['2013':].copy()

GB_model = GradientBoostingRegressor()
GB_model.fit(traindata[list_feat], traindata['kWh'])
testdata['kWh_pred'] = GB_model.predict(testdata[list_feat])
errors = abs(testdata['kWh_pred'] - testdata['kWh'])
RSQUARED = r2_score(testdata['kWh'], testdata['kWh_pred'])
MAE = mean_absolute_error(testdata['kWh'], testdata['kWh_pred'])

print("Model: GradientBoostingRegressor")
print("R SQUARED: "+str(round(RSQUARED,4)))
print("MAE: "+str(round(MAE,4)))

plt.figure(figsize=(8,2))
plt.plot(testdata.iloc[:7*24].index, testdata.iloc[:7*24]['kWh'], '-k', alpha=0.6, label='kWh')
plt.plot(testdata.iloc[:7*24].index, testdata.iloc[:7*24]['kWh_pred'], '--g', alpha=0.8, label='kWh_pred')
plt.legend(loc='lower center', ncol=3, bbox_to_anchor=(0.5, -0.5))
plt.show() 
```

Again we see reasonable forecast accuracy, but with the same over-forecast during the morning hour. Is there a way to fix this issue?

### The effect of aggregation size

Besides the effects of prediction horizon, the aggregation level is another factor influencing the forecasting model's performance. So, here we compare the forecasting results from 10 to 1000 houses.

```python
list_n_sample = []
GB_RSQUARED = []
GB_MAE = []

for n_sample in tqdm.tqdm(np.arange(10,1000,10)):
    elec_houses = elec_house.sample(n=n_sample, random_state=random_state, axis=1).mean(axis=1).to_frame()
    elec_houses.columns = ['kWh']

    df_dataset = elec_houses.copy()

    df_dataset['timestamp'] = df_dataset.index
    df_dataset['date'] = pd.to_datetime(df_dataset.index.date)
    df_dataset['year'] = df_dataset.index.year.astype('int')
    df_dataset['weekday'] = df_dataset.index.weekday
    df_dataset['hour'] = df_dataset.index.hour + df_dataset.index.minute/60
    df_dataset['timeofweek'] = df_dataset['hour'] + df_dataset['weekday']*24

    df_dataset = df_dataset.merge(weather_hourly_darksky[['temperature']], left_index=True, right_index=True)
    df_dataset['kWh_yesterday'] = df_dataset['kWh'].shift(24)
    df_dataset['kWh_lastWeek'] = df_dataset['kWh'].shift(7*24)


    df_dataset.set_index('timestamp',inplace=True)

    df_dataset = df_dataset.dropna()    
    
    traindata = df_dataset.loc['2013-01':'2013-06'].copy()
    testdata = df_dataset.loc['2013-07':].copy()

    GB_model = GradientBoostingRegressor()
    GB_model.fit(traindata[list_feat], traindata['kWh'])
    testdata['kWh_pred'] = GB_model.predict(testdata[list_feat])
    RSQUARED = r2_score(testdata['kWh'], testdata['kWh_pred'])
    MAE = mean_absolute_error(testdata['kWh'], testdata['kWh_pred'])
    GB_RSQUARED.append(RSQUARED)
    GB_MAE.append(MAE)    
    
    list_n_sample.append(n_sample)  
```

In terms of R-squared and MAE, the energy consumption with increasing aggregation houses has better predictabilty than single or few houses.

```python
df_RSQUARED = pd.DataFrame(index=list_n_sample, data={'Gradient Boosting': GB_RSQUARED})
df_RSQUARED.plot(figsize=(5,3))
plt.xlabel('Number of houses for aggregation')
plt.ylabel('R-Squared')
plt.show()
df_MAE = pd.DataFrame(index=list_n_sample, data={'Gradient Boosting': GB_MAE})
df_MAE.plot(figsize=(5,3))
plt.xlabel('Number of houses for aggregation')
plt.ylabel('MAE')
plt.show()
```

### How much data are needed?

Besides the feature enegineering as we build forecasting models, the size of train data also matters. Here we train the model by ranging from last 30 days to 400 days to test the performance of the forecasting accuracy.

```python
list_feat = ['weekday', 'hour','temperature', 'kWh_lastWeek','kWh_yesterday']
```

```python
days_of_train_data = []
r2_metric = []
mae_metric = []

for days in tqdm.tqdm(np.arange(30, 400, 10)):
    traindata = df_dataset.loc[:'2012'].copy()
    traindata = traindata.iloc[-days*24:]
    testdata = df_dataset.loc['2013':].copy()

    GB_model = GradientBoostingRegressor()
    GB_model.fit(traindata[list_feat], traindata['kWh'])
    testdata['kWh_pred'] = GB_model.predict(testdata[list_feat])
    errors = abs(testdata['kWh_pred'] - testdata['kWh'])
    RSQUARED = r2_score(testdata['kWh'], testdata['kWh_pred'])
    MAE = mean_absolute_error(testdata['kWh'], testdata['kWh_pred'])

    days_of_train_data.append(days)
    r2_metric.append(RSQUARED)
    mae_metric.append(MAE)
```

Generally speaking, the size of train data matters to the forecasting accuracy. At least three or four months of train data could help the model achieve a high performance with ~0.96 R-squared. However, the interesting fact is that more data also might slightly harm the forecasting performance as well, especially when the size is more than one year.

```python
df_metrics = pd.DataFrame([days_of_train_data,r2_metric,mae_metric],index=['Days of train data','RSQUARED','MAE']).T
df_metrics.plot(kind='scatter',x='Days of train data',y='RSQUARED', figsize=(5,3))
plt.show()
df_metrics.plot(kind='scatter',x='Days of train data',y='MAE', figsize=(5,3))
plt.show()
```
